### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

# Ingresos totales de cada tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total de la tienda 1: {ingreso_tienda1}")
print(f"Ingreso total de la tienda 2: {ingreso_tienda2}")
print(f"Ingreso total de la tienda 3: {ingreso_tienda3}")
print(f"Ingreso total de la tienda 4: {ingreso_tienda4}")
```

```python
# prompt: genera la grafica

import matplotlib.pyplot as plt

# Datos de ingresos (reemplaza con tus valores reales)
ingresos = {
    'Tienda 1': ingreso_tienda1,
    'Tienda 2': ingreso_tienda2,
    'Tienda 3': ingreso_tienda3,
    'Tienda 4': ingreso_tienda4
}

# Crea la gráfica de barras
tiendas = list(ingresos.keys())
valores = list(ingresos.values())

plt.figure(figsize=(10, 6))  # Ajusta el tamaño de la figura si es necesario
plt.bar(tiendas, valores, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.xlabel("Tiendas")
plt.ylabel("Ingresos totales")
plt.title("Ingresos totales por tienda")

# Agrega etiquetas de valor sobre cada barra
for i, v in enumerate(valores):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

# 2. Ventas por categoría

```python
# Creamos un diccionario vacío para guardar los resultados
productos_por_categoria = {}

# Recorremos cada tienda por separado
for tienda_df in [tienda, tienda2, tienda3, tienda4]:
    nombre_tienda = tienda_df["tienda"].iloc[0]  # Obtenemos el nombre de la tienda
    categorias = tienda_df["Categoría del Producto"].value_counts()  # Contamos los productos por categoría

    # Guardamos el resultado en el diccionario
    productos_por_categoria[nombre_tienda] = categorias

# Mostramos los resultados
for tienda, categorias in productos_por_categoria.items():
    print(f"\nProductos vendidos por categoría en {tienda}:")
    print(categorias)


```

# 3. Calificación promedio de la tienda


```python
# Cargar los datos de cada tienda
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Agregar una columna para identificar la tienda
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Juntar todas las tiendas en una sola tabla
todas_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])

# Crear una lista para guardar los resultados
resultados = []

# Lista de nombres de las tiendas
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Calcular el promedio de calificaciones para cada tienda
for tienda in nombres_tiendas:
    # Filtrar los datos de la tienda actual
    datos_tienda = todas_tiendas[todas_tiendas['Tienda'] == tienda]

    # Sumar todas las calificaciones
    suma_calificaciones = 0
    conteo = 0
    for calificacion in datos_tienda['Calificación']:
        if pd.notna(calificacion):  # Verificar que la calificación no sea nula
            suma_calificaciones += calificacion
            conteo += 1

    # Calcular el promedio (si hay datos)
    if conteo > 0:
        promedio = suma_calificaciones / conteo
        promedio = round(promedio, 2)  # Redondear a 2 decimales
    else:
        promedio = 0  # Si no hay datos, el promedio es 0

    # Guardar el resultado
    resultados.append([tienda, promedio])

# Crear una tabla con los resultados
tabla_resultados = pd.DataFrame(resultados, columns=['Tienda', 'Calificación Promedio'])

# Mostrar la tabla
print("Calificaciones Promedio por Tienda:")
print(tabla_resultados)

# Crear un gráfico de pastel
tiendas = tabla_resultados['Tienda']
calificaciones = tabla_resultados['Calificación Promedio']

# Colores para cada tienda
colores = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow']

# Hacer el gráfico de pastel
plt.pie(calificaciones, labels=tiendas, colors=colores, autopct='%.2f')

# Hacer que el pastel sea redondo
plt.axis('equal')

# Título del gráfico
plt.title('Calificación Promedio por Tienda')

# Mostrar el gráfico
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Nombres de las tiendas
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
dataframes = [tienda1, tienda2, tienda3, tienda4]

# Recorremos cada tienda
for nombre_tienda, tienda_df in zip(nombres_tiendas, dataframes):
    # Contamos cuántas veces se vendió cada producto
    productos_vendidos = tienda_df["Producto"].value_counts()

    # Tomamos los 5 más vendidos y los 5 menos vendidos
    mas_vendidos = productos_vendidos.head(5)
    menos_vendidos = productos_vendidos.tail(5)

    # Visualización
    fig, axs = plt.subplots(1, 2, figsize=(14, 5))
    fig.suptitle(f"Productos más y menos vendidos en {nombre_tienda}", fontsize=16)

    # Más vendidos (gráfico de línea)
    axs[0].plot(mas_vendidos.index, mas_vendidos.values, marker='o', color='green', linestyle='-', linewidth=2)
    axs[0].set_title("Más vendidos")
    axs[0].set_ylabel("Cantidad vendida")
    axs[0].tick_params(axis='x', rotation=45)

    # Menos vendidos (gráfico de línea)
    axs[1].plot(menos_vendidos.index, menos_vendidos.values, marker='o', color='red', linestyle='-', linewidth=2)
    axs[1].set_title("Menos vendidos")
    axs[1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()


```

# 5. Envío promedio por tienda

```python
# Nombres de las tiendas y sus DataFrames
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
dataframes = [tienda, tienda2, tienda3, tienda4]

# Diccionario para guardar los promedios
costo_envio_promedio = {}

# Calculamos el promedio de envío por tienda
for nombre_tienda, tienda_df in zip(nombres_tiendas, dataframes):
    promedio = tienda_df["Costo de envío"].mean()
    costo_envio_promedio[nombre_tienda] = promedio

# Mostramos los resultados
print("Costo de envío promedio por tienda:\n")
for tienda, promedio in costo_envio_promedio.items():
    print(f"{tienda}: ${promedio:.2f}")
```
